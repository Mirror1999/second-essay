## 在加入一个对疫情的衡量，可以根据不同阶段进行划分
## 以上结果说明，要对数据进行进一步分段，结果如下：
- 根据每日新增的情况，将时间段划分为6个
- 1 2019年6/1到2020/1/7：未感染阶段，level为0
- 2 2019/1/8到2020/4/22：感染初期到爆发  level为5
- 3 2020/4/23到2021/11/09：疫情得到控制   level为4
- 5 2020/4/22到2021/5/17：疫情常态化     level为3
- 6 2022/5/18到2022/8/4：疫情得到控制     level为3
- 7 2022/8/5到2022/11/23：疫情复发       level为4

```python
def level_from_period(int1):
    int1 = int(int1)
    if int1==1:
        return 0
    
    elif int1==2:
        return 5
    
    elif int1==3:
        return 4
    
    elif int1==5:
        return 3
    
    elif int1==6:
        return 3
    
    elif int1==7:
        return 4
    else:
        return False
df_new_scenceName_sentimentScore_weather2['level'] = df_new_scenceName_sentimentScore_weather2['period'].apply(level_from_period)
```

```python
def split_other_timestring(string1):
    string1 = string1.replace(' ','')
    list_string2 = string1.split('/')
    string2 = list_string2[0]+'-'+list_string2[1]+'-'+list_string2[2]
    string2 = string2.replace(' ','')
    return string2
file1['comment_day1'] = file1['date'].apply(split_other_timestring)
file1['comment_day'] = file1['comment_day1'].apply(string_to_date)
```

```python
# 打开固定效应.csv
import pandas as pd
file2 = pd.read_csv('固定效应.csv',encoding = 'utf-8')
type(file2['comment_day'][1])
file2
```

```python
from datetime import datetime
def string_to_date(date_string1):
    date_string1 = date_string1.replace(' ','')
    date_object1 = datetime.strptime(date_string1, '%Y-%m-%d')
    return date_object1
file2['comment_day'] = file2['comment_day'].apply(string_to_date)
file2
```

```python
# 将file1 的date 与 file2 的 comment_day相连
df_add_confirmed = file2.merge(file1,on='comment_day',how='left')
df_add_confirmed.fillna(0)
df_add_confirmed
```

```python
# 提取需要的属性，组成df_sentiment_confirmed
# comment_day	scence_name  sentiment_score	scence_id	city_new	high_temperature	holiday	outbreak_timeDistance	outbreak	period	Month	level	
df_sentiment_confirmed = df_add_confirmed[['comment_day','scence_id','scence_name','sentiment_score','scence_id','city_new','high_temperature','holiday','outbreak','outbreak_timeDistance','period','Month','level','confirmed']]
df_sentiment_confirmed
```

```python
df_sentiment_confirmed = df_sentiment_confirmed.drop_duplicates(('comment_day','scence_name'))
df_sentiment_confirmed = df_sentiment_confirmed.fillna(0)
type(df_sentiment_confirmed)
```

```python
def tenwords(string1):
    string1 = str(string1)
    string2 = string1[:11]
    string2 = string2.replace(' ','')
    return string2
df_sentiment_confirmed['comment_day'] = df_sentiment_confirmed['comment_day'].apply(tenwords)
df_sentiment_confirmed['comment_day'][2]
```

```python
from statsmodels.datasets import grunfeld
# # data = grunfeld.load_pandas().data
data = df_sentiment_confirmed.set_index(['scence_name','comment_day'])
# print(data.head())
```

```python
# import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
x = df_sentiment_confirmed['outbreak']
y = df_sentiment_confirmed['sentiment_score']
x= sm.add_constant(x) #添加常数项
est=sm.OLS(y,x)
model=est.fit()#建立最小二乘回归模型
print(model.summary())
# sentiment(it) = 44.6127 -1.3158*outbreak(it),且outbreak对应的p值小于0.05***
```
