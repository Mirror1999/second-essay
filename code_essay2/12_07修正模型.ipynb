```python
import pymysql
import pandas as pd
#连接数据库
conn=pymysql.connect(host = '127.0.0.1' # 连接名称，默认127.0.0.1
,user = 'root' # 用户名
,passwd='123456' # 密码
,port= 3306 # 端口，默认为3306
,db='weibo' # 数据库名称
,charset='utf8' # 字符编码
)
cur = conn.cursor() # 生成游标对象
sql="select * from `post`  " # SQL语句
cur.execute(sql) # 执行SQL语句
data = cur.fetchall() # 通过fetchall方法获得数据
for i in data[:20]: # 打印输出前3条数据
    print(i)
# cur.close() # 关闭游标
# conn.close() # 关闭连接
```

```python
# # 选取前十个调试
# data_10 = data[:10]
# 全部数据
data_10 = data
import pandas as pd
# list_data = list(data_10)
def tuple_to_list(t1):
    return list(t1)
list_data = list(map(tuple_to_list,data_10))
#list_data
```

```python
def split_data(list1):
    string1 = str(list1)
    list2 = string1.split(',')
    return list2
    
data_list_all = list(map(split_data,list_data))
#data_list_all
```

```python
#### - comment_id: 0
#### - scence_name:1
#### - user_name:2
#### - comment_time:3
#### - comment_detail:4
#### - crawl_time:-1
# 将list换分为新list
list_comment_id = []
list_scence_name = []
list_user_name = []
list_comment_time = []
list_comment_detail = []
list_crawl_time = []

for list1 in data_list_all:
    string0 = str(list1[0]).replace('[','')
    string0 = int(string0)
    # 清洗comment_id数据,并转换为整数类型
    
    string1 = str(list1[1]).replace('\'','')
    string1 = string1.strip()
    # 清洗scence_name数据
    
    string2 = str(list1[2]).replace('\'','')
    string2 = string2.strip()
    # 清洗user_name数据   
    
    string3 = str(list1[3]).replace('\'','')
    string3 = string3.strip()
    # 清洗comment_time数据       
    
    string4 = str(list1[4]).replace('\'','')
    string4= string4.replace('<br>',' ')
    string4 = string4.strip()
    # 清洗comment_detail数据                                
    
    string_1 = str(list1[-1]).replace('\'','')
    string_1 = string_1.replace(']',' ')
    # 清洗crawl_time数据                                
        
    
    
    list_comment_id.append(string0) 
    list_scence_name.append(string1) 
    list_user_name.append(string2) 
    list_comment_time.append(string3)
    list_comment_detail.append(string4) 
    list_crawl_time.append(string_1) 
# 将数据转换为dataframework类型
# 构建字典
dic_new = {'comment_id':list_comment_id,
           'scence_name':list_scence_name,
           'user_name':list_user_name,
           'comment_time':list_comment_time,
           'comment_detail':list_comment_detail,
           'crawl_time':list_crawl_time}
data_new = pd.DataFrame(dic_new)
data_new
```

```python
#计算
import json
from tencentcloud.common import credential
from tencentcloud.common.profile.client_profile import ClientProfile
from tencentcloud.common.profile.http_profile import HttpProfile
from tencentcloud.common.exception.tencent_cloud_sdk_exception import TencentCloudSDKException
from tencentcloud.nlp.v20190408 import nlp_client, models
list_median_sentiment = []
try:
    # 实例化一个认证对象，入参需要传入腾讯云账户secretId，secretKey,此处还需注意密钥对的保密
    # 密钥可前往https://console.cloud.tencent.com/cam/capi网站进行获取
    cred = credential.Credential("AKIDqVxB8oDMHwlLm6Bk23oCZovpVw71BBzX", "MSItEJAnBPO8d3GEwPedOu99vSEk14qM")

# SecretId: AKIDqVxB8oDMHwlLm6Bk23oCZovpVw71BBzX
# SecretKey: MSItEJAnBPO8d3GEwPedOu99vSEk14qM    
    for text1 in list_texts_all:
    # 实例化一个http选项，可选的，没有特殊需求可以跳过
        httpProfile = HttpProfile()
        httpProfile.endpoint = "nlp.tencentcloudapi.com"

    # 实例化一个client选项，可选的，没有特殊需求可以跳过
        clientProfile = ClientProfile()
        clientProfile.httpProfile = httpProfile
    # 实例化要请求产品的client对象,clientProfile是可选的
        client = nlp_client.NlpClient(cred, "ap-guangzhou", clientProfile)
    
    
        

    # 实例化一个请求对象,每个接口都会对应一个request对象
        req = models.SentimentAnalysisRequest()
        params = {
         "Text": text1,
         "Mode": "3class"
    }
        req.from_json_string(json.dumps(params))

    # 返回的resp是一个SentimentAnalysisResponse的实例，与请求对象对应
        resp = client.SentimentAnalysis(req)
        list_median_sentiment.append(resp.to_json_string())
    # 输出json格式的字符串回包
        print(resp.to_json_string())
except TencentCloudSDKException as err:
    print(err)

print('all words MORE than 200 in compete!')
```

```python
# 将list_median_sentimen进行拆分
# dic_list_median_sentiment = list_median_sentiment[1]
def string_dic(string1):
    string1 = string1.replace('{','')
    string1 = string1.replace('}','')
    string1 = string1.replace('"','')
    dic_item_median_sentiment = dict(x.split(":") for x in string1.split(","))
    return dic_item_median_sentiment
# 
```

```python
# string_dic(list_median_sentiment)
# list_median_sentiment1 = list(map(string_dic,list_median_sentiment))
# list_median_sentiment1
list_median_new = []

# {"Positive": 0.36061561, "Neutral": 0.53547966, "Negative": 0.10393469, "Sentiment": "neutral", "RequestId": "dbc96b0e-a81d-4b5b-89d6-f41483e567b7"}


list_median_Positive = []
list_median_Neutral = []
list_median_Negative = []
list_median_Sentiment = []
list_median_RequestId = []

for item1 in list_median_sentiment:
    list_values = []
    item_median = string_dic(item1)
    list_values = list(item_median.values())
    list_median_Positive.append(float(list_values[0])*100)
    list_median_Neutral.append(float(list_values[1])*100)
    list_median_Negative.append(float(list_values[2])*100)
    list_median_Sentiment.append(list_values[3])
    list_median_RequestId.append(list_values[4])
    
    # list_median_new.append(item_median)
print(len(list_median_Positive))
print(len(list_median_Neutral))
print(len(list_median_Negative))
print(len(list_median_Sentiment))
print(len(list_median_RequestId))
```
