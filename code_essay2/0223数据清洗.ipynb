```python
from datetime import datetime
def string_to_date(string1):
    string1 = string1.strip()
    date_object1 = datetime.strptime(string1, '%Y-%m-%d')
    return date_object1

date_holiday['date'] = date_holiday['date'].apply(string_to_date)
```

```python
# 对时间段进行划分
def period_segment(string1):
    string1 = str(string1)
    if string1<'2020-01-23':
        period = 0
    elif string1<'2020-04-22':
        period = 1
    elif string1<'2020-11-09':
        period = 2
    elif string1<'2021-05-17':
        period = 3
    elif string1<'2022-08-04':
        period = 4
    elif string1<'2022-12-06':
        period = 5  
    else:
        period = 6      
    return period
df8['period'] =  df8['comment_day'].apply(period_segment)
df8
```
